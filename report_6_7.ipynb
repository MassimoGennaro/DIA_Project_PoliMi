## Combining Advertising and Pricing campaign

In this setting we combine the algorithms of **Advertising** and **Pricing**, trying to decide the allocation of budget in the different subcampaigns using in the knapsack-like optimization the **product of number of clicks and expected rewards** as _value-per-click_.

The second term is computed at each round of the budget allocation by the pricing algorithm, that uses as users the number of people (clicks) given by the expected values of the learners of the advertising algorithm. 

The two problems can be decompose since we assume that the classes of users in the pricing algorithm are known and each subcampaign targets a single class of users.
